```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Count the number of students for each exam score
exam_score=df1.groupby('exam_score')['student_id'].count()
exam_score
```

```python
# Plot distribution of exam scores
exam_score=df1.groupby('exam_score')['student_id'].count()
exam_score.plot(kind='bar',color='purple')
plt.title('Distribution of exam_score')
plt.xlabel('student id')
plt.ylabel('exam score')
plt.legend()
plt.show()
```

```python
# Plot exam scores of each student
plt.bar(df1["student_id"],df1["exam_score"],color='blue')
plt.xlabel("Student ID")
plt.ylabel("Exam Score")
plt.title("Exam Score of Each Student")
plt.xticks(rotation=90)
plt.show()
```

```python
# Get sorted count of each exam score
exam_score_dist=df1["exam_score"].value_counts().sort_index()
exam_score_dist
```

```python
# Plot average exam score vs hours studied
hours_group=df1.groupby("hours_studied")["exam_score"].mean()
hours_group.plot(kind="bar", color="purple")
plt.xlabel("Hours Studied")
plt.ylabel("Average Exam Score")
plt.title("Study Hours vs Exam Score")
plt.show()
```

```python
# Plot average exam score vs attendance percentage
attendance_group=df1.groupby("attendance_percent")["exam_score"].mean()
attendance_group.plot(kind="bar", color="purple")
plt.xlabel("Attendance")
plt.ylabel("Average Exam Score")
plt.title("Attendance vs Exam Score")
plt.show()
```

```python
# Plot average exam score vs sleep hours
sleep_group=df1.groupby("sleep_hours")["exam_score"].mean()
sleep_group.plot(kind="bar", color="purple")
plt.xlabel("Sleep Hours")
plt.ylabel("Average Exam Score")
plt.title("Sleep Hours vs Exam Score")
plt.show()

```

```python
# Plot average exam score vs previous exam scores
prev_group=df1.groupby("previous_scores")["exam_score"].mean()
prev_group.plot(kind="bar",color="purple")
plt.xlabel("Previous Scores")
plt.ylabel("Average Exam Score")
plt.title("Previous Scores vs Exam Score")
plt.show()
```

```python
# Drop student_id column and compute correlation between features
df=df1.drop(columns=['student_id'])
correlation=df.corr()
correlation
```

```python
# Plot correlation of each feature with exam score
df2=df1.drop(columns=["student_id"])
df2
correlation=df2.corr()["exam_score"].drop("exam_score")
correlation.plot(kind='bar', color=['purple','black','green','yellow'])
plt.title("Correlation of Features with Exam Score")
plt.xlabel("Features")
plt.ylabel("Correlation Value")
plt.show()
```
